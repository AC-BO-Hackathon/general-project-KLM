```python
# def featurize(df, formula_col='formula', pbar=False, n_jobs=None, n_chunksize=None):
#     """
#     Extract magpie feature set from formula
#     """
#     starttime = time.time()
#     if formula_col not in list(df):
#         raise KeyError(f"Data does not contain {formula_col} column")
#     print("Featurizing dataset...")

#     if n_jobs:
#         stc.set_n_jobs(n_jobs)

#     stc = StrToComposition()

#     feat = stc.featurize_dataframe(df, formula_col, ignore_errors=True, pbar=pbar)
#     element_property = ElementProperty.from_preset(preset_name='magpie')
#     for f in [element_property, BandCenter(), AtomicOrbitals()]:
#         if n_jobs:
#             f.set_n_jobs(n_jobs)
#         if n_chunksize:
#             f.set_chunksize(n_chunksize)
#         feat = f.featurize_dataframe(feat, "composition", pbar=pbar, ignore_errors=True)

#     feat = feat.set_index(feat[formula_col])
#     print(f"Featurization time: {round((time.time() - starttime) / 60, 2)} min")
#     return feat


def get_identical_cols(df):
    """
    Get the columns of a dataframe which have identical values.
    """
    xx = {}
    for c in df:
        vals1 = list(df[c])
        for c2 in df:
            vals2 = list(df[c2])
            # if the columns contain identical values
            if c != c2 and vals1 == vals2:
                # if these values have not been seen before
                if vals1 not in [list(df[y]) for y in xx]:
                    xx[c] = [c2]
                # if these values have been seen before
                else:
                    y = [yy for yy in xx if list(df[yy]) == vals1][0]
                    if c2 != y and c2 not in xx[y]:
                        xx[y].append(c2)
    return xx


def process_dataset(df, target):
    """
    Clean dataset and extract magpie features if formula is provided
    """
    # drop extreme outlier rows?

    # featurize magpie datasets
    # if 'formula' in df.columns:
    #     df = featurize(df)

    # remove non-numeric columns
    df = df[[c for c in df if pd.api.types.is_numeric_dtype(df[c])]]
    # remove columns with not many unique values?

    # remove constant columns
    df = df.loc[:, (df != df.iloc[0]).any()]

    # change all infinite values to nan
    df = df.replace([np.inf, -np.inf], np.nan)

    # remove rows containing a nan
    df = df.dropna()

    # remove columns with identical values
    remove_cols = [v0 for _, v in get_identical_cols(df).items() for v0 in v]
    df = df[[c for c in df if c not in remove_cols]]

    # for some datasets, each input feature x could have been evaluated more than once.
    # take average
    features = set(df.columns)
    features.remove(target)
    df = df.groupby(list(features))[target].agg(lambda x: x.unique().mean())
    df = (df.to_frame()).reset_index()
    return df
```

# Feature Engineering Functions

```python
def array_is_constant(arr: list) -> np.array:
    """Check if a 1D array is constant"""
    return np.allclose(arr, np.repeat(arr[0], len(arr)))


def add_additional_cols(
    df: pd.DataFrame,
    ignore: list = [],
    nat_log: bool = True,
    powers: list = [-4, -3, -2, -1, -0.5, -0.333, -0.25, 0.25, 0.33, 0.5, 2, 3, 4],
) -> pd.DataFrame:
    """
    Add additional columns to a dataframe by executing
    mathematical operations on existing columns.
    """
    # now raise existing columns to varying powers
    new_col_names = []
    new_vals_all = np.empty((len(df), 0))

    # loop over each column to use for creating additional columns
    for c in [cc for cc in df if cc not in ignore]:

        # get original column values
        vv = df[c].values

        # raise existing columns to various powers
        for p in powers:
            # first assess mathematical viability of different conditions.
            # for example, we can't perform (-2)^(1/2).
            # if conditions are met, save new column
            if any(
                [
                    p > 0 and isinstance(p, int),  # pos integer powers
                    p < 0 and isinstance(p, int) and 0 not in vv,  # neg integer powers
                    p > 0
                    and not isinstance(p, int)
                    and np.all(vv >= 0),  # pos non-int powers
                    p < 0
                    and not isinstance(p, int)
                    and np.all(vv > 0),  # neg non-int powers
                ]
            ):

                new_col_vals = np.float_power(vv, p).reshape((-1, 1))
                if (
                    not array_is_constant(new_col_vals)
                    and np.isfinite(new_col_vals).all()
                ):
                    new_vals_all = np.hstack((new_vals_all, new_col_vals))
                    new_col_names.append(f"{c}**{str(p)}")

        # take natural logs of existing columns
        if nat_log:
            if not c.startswith("ln ") and np.all(vv > 0):
                new_col_vals = np.log(vv).reshape((-1, 1))
                if (
                    not array_is_constant(new_col_vals)
                    and np.isfinite(new_col_vals).all()
                ):
                    new_vals_all = np.hstack((new_vals_all, new_col_vals))
                    new_col_names.append(f"{c}**ln")

    # combine new columns with original columns in a single dataframe
    new_df = pd.DataFrame(data=new_vals_all, columns=new_col_names, index=df.index)
    df = pd.concat([df, new_df], axis=1)
    return df


def find_top_n_features(df, original_features, target, n):
    """
    Finds the top n features of the dataset according to how well they fit
    to a quadratic polynomial.

    Note: A diversity criteria has been enforced to prevent the top features
    being powers of the same feature when adding additional columns (ie. we are
    trying to prevent the case where the top features are: [temp, temp**2, temp**3...])
    """
    # obtain a list of all features
    features = set(df.columns)
    features.remove(target)
    features = list(features)

    # create dictionary where the keys are the original features and the values
    # are a list of tuples of the form: (r2, features based on original feature)
    # eg. {'temp': [(0.9, temp), (0.87, temp**2), (0.95, temp**3)...]}
    unsorted_features = {f: [] for f in original_features}
    for col in features:
        # find how well the feature fits to the target using a quadratic polynomial
        fit = np.polyfit(df[col], df[target], 2, full=True)
        ssr = fit[1] # sum of squared residuals
        sst = np.sum((df[target] - df[target].mean())**2) # total sum of squares

        # if ssr is 0, numpy will return a blank list
        if len(ssr) == 0:
            # if the number of unique values is less than 4, the fit will always
            # be perfect. Therefore, set ssr to sst and force r2 to be 0
            if len(df[col].unique()) < 4:
                ssr = sst
            else:
                ssr = 0
        else:
            ssr = ssr[0]
        r2 = 1 - ssr / sst

        original_col = col.split('**')[0]
        unsorted_features[original_col].append((r2, col))

    # sort features based on r2 value
    # eg. {'temp': [(0.95, temp**3), (0.9, temp), (0.87, temp**2)...]}
    sorted_features = {
        f: sorted(unsorted_features[f])[::-1] for f in original_features
    }

    # find top features while enforcing diversity criteria
    top_n = []
    for i in range(n):
        # create mini list of top "diverse" features
        # eg. [(0.98, temp**3), (0.95, theta**0.5), (0.99, t**3)...]
        top_i = []
        for original_col in sorted_features:
            if i < len(sorted_features[original_col]):
                top_i.append(sorted_features[original_col][i])

        # sort mini list of top "diverse" features
        # eg. [(0.99, t**3), (0.98, temp**3), (0.95, theta**0.5)...]
        top_i = sorted(top_i)[::-1]
        for j in top_i:
            top_n.append(j[1])
            if len(top_n) >= n:
                break
        # once n features have been found, break out of loop
        if len(top_n) >= n:
            break

    return top_n
```

# Acquisition Functions

```python
def EI(mean, std, y_best, ratio):
    std = ratio * std

    z = (y_best - mean)/std
    return (y_best - mean) * norm.cdf(z) + std * norm.pdf(z)
```

# Run Framework

```python
# framework


# good practice to keep check of time used
start_time = time.time()

# these will carry results along optimization sequence from all n_ensemble runs
index_collection = []
X_collection = []
y_collection = []
TopCount_collection = []

#keep track of longest learning cycle
longest_learning_cycle = 0


for s in seed_list:

    if len(index_collection) == n_ensemble:
        break

    print('initializing seed = ' +str(seed_list.index(s)))
    random.seed(s)

    indices = list(np.arange(N))
# index_learn is the pool of candidates to be examined
    index_learn = indices.copy()
# index_ is the list of candidates we have already observed
#     adding in the initial experiments
    index_ = random.sample(index_learn, n_initial)

#     list to store all observed good candidates' input feature X
    X_ = []
#     list to store all observed good candidates' objective value y
    y_ = []
#     number of top candidates found so far
    c = 0
#     list of cumulative number of top candidates found at each learning cycle
    TopCount_ = []
#     add the first n_initial experiments to collection
    for i in index_:
        X_.append(X_feature[i])
        y_.append(y[i])
        if i in top_indices:
            c += 1
        TopCount_.append(c)
        index_learn.remove(i)


#     for each of the the rest of (N - n_initial) learning cycles
#     this for loop ends when all candidates in pool are observed

    lawrence = 0
    for i in np.arange(len(index_learn)):
        print(lawrence)
        lawrence += 1
        y_best = np.min(y_)

        s_scaler = preprocessing.StandardScaler()
        X_train = s_scaler.fit_transform(X_)
        y_train = s_scaler.fit_transform([[i] for i in y_])
        y_train = np.ravel(y_train)

        if combo['model'] == 'RF':
            n_est = 50
            model = RandomForestRegressor(n_estimators= n_est, n_jobs= -1)
            model.fit(X_train, y_train)
        else:
            n_est = 5
            model = BaggingRegressor(LinearRegression(),n_estimators = n_est, n_jobs = -1)
            model.fit(X_train, y_train)


#         by evaluating acquisition function values at candidates remaining in pool
#         we choose candidate with larger acquisition function value to be observed next
        next_index = None
        max_ac = -10**10
        for j in index_learn:
            X_j = X_feature[j]
            y_j = y[j]

            # TODO: CREATE PREDICTION FUNCTION WITH UNCERTAINTY
            ensemble_predictions = []
            for k in np.arange(n_est):
                ensemble_predictions.append((model.estimators_[k].predict(np.array([X_j]))).tolist())
            mean = np.mean(np.array(ensemble_predictions), axis=0)[0]
            std = np.std(np.array(ensemble_predictions), axis=0)[0]

            # select acquisition function
            ac_value = EI(mean, std, y_best, combo['acq_func'])

            if max_ac <= ac_value:
                max_ac = ac_value
                next_index = j



        X_.append(X_feature[next_index])
        y_.append(y[next_index])


        if next_index in top_indices:
            c += 1
        print('num top: ', c)

        TopCount_.append(c)

        index_learn.remove(next_index)
        index_.append(next_index)
        if c == len(top_indices):
            break

    if len(TopCount_)>longest_learning_cycle:
        longest_learning_cycle = len(TopCount_)    
        

    # assert len(index_) == N

    index_collection.append(index_)
    X_collection.append(X_)
    y_collection.append(y_)
    TopCount_collection.append(TopCount_)








    print('Finished seed')

total_time = time.time() - start_time
print(total_time)
TopCount_collection # PUT INTO DATASET


# master = np.array([index_collection, X_collection, y_collection, TopCount_collection, total_time])
# #  #TODO: name output file
# np.save(f'{combo["dataset"]} ({combo["n_features"]}, {combo["acq_func"]}, {combo["model"]})', TopCount_collection)
```
